In Part-A, I have done all the requirements as an application of message delivery.

1. Printing a Welcome Message to the user.

2. Taking a string and a letter from the user.

3. Counting the number of occurences of the letters in the message.

4. Calculating the percentage of letters used in the message.

The reason I used the Object Oriented Programming is; the above given requirement best holds as a data structure, which I have done in a form of UTA message delivery application below which satisfies all the requirements of this mini project.

```python
class word_summary:

  def __init__(self):
    self.word = ""
    self.number_source = ""
    self.number_destination = ""
    self.message = """"""
    self.freq = {}
    self.prop = {}

  def input_details(self):

    print("UTA Message Delivery System:")
    print("_"*80)
    self.word= input('Enter your name and your Name: ')
    print("_"*80)
    self.number_source = int(input('Enter your phone Number: '))
    print("_"*80)
    self.number_destination = int(input('Enter the destination phone Number: '))
    print("_"*80)
    print(f'Hi {self.word}, Welcome to UTA Message Delivery System.')
    print("_"*80)
    print(f'Your Phone Number is: {self.number_source}')
    print("*"*80)
    print(f'Your Destination Number is: {self.number_destination}')
    print("*"*80)
    self.message = input('Enter your message: ').strip().lower()
  def letter_frequency(self):
    for i in self.message:
      if i in self.freq:
        self.freq[i] += 1
      else:
        self.freq[i] = 1
    print(self.freq)
  def letter_proportion(self):
    if self.freq == {}:
       self.letter_frequency()
    for i in self.freq:
      self.prop[i] = (self.freq[i]/len(self.message))*100
    print(self.prop)
```

```python
snowball = """
snowball is one of the most famous poems in literature. It says.
I made myself a snowball.
As perfect as could be.
I thought I'd keep it as a pet.
And let it sleep with me.
I made it some pajamas.
And a pillow for its head.
Then last night it ran away,
But first it wet the bed
"""
```

```python

#This method is an organic string method with Ignore Case condition.
count = 0
for index, letter in enumerate(snowball.lower()):
  if letter == 'i':
    if snowball[index+1] == 't':
      count += 1
print(f'The number of occurances of the word \'it\' in the string snowball : {count}')

```

```python
# To find the number of occurences of 'it' without ignoring cases.
# We can either create a pattern 'it' or use character classes [i][t]

counts = re.findall(r'[i][t]', snowball)
it_count = len(counts)
print(len_counts)
```

```python

#This method is an organic string method that doesn't include the Ignore case condition.
count = 0
for index, letter in enumerate(snowball):
  if letter == 'i':
    if snowball[index+1] == 't':
      count += 1
print(f'The number of occurances of the word \'it\' in the string snowball : {count}')

```

```python
#2.
# There are lowecase letters, uppercase letters, punctuations like(" , ", " . ", " ' ") white spaces and few newline
# characters in the above snowball string.
# This is a clear and organic regular expression approach
len(re.findall(r"[a-zA-Z0-9'\s,\.\n]", snowball))
```

```python
# We can also do it in a approach with whitespace and non whitespace characters.(\s, \S)
# but technically this isn't a clear approach
len(re.findall(r"[\s\S]", snowball))
```

```python
#This approach is an organic string approach without using any regex.
# Here we initialize a list of punctuations
# Iterate the string within the requiremetns in the question to check whitespaces, lowercase,
# Upper case, punctuations.
# List of common punctuation marks
punctuations = [
    '.',       # Period
    ',',       # Comma
    '?',       # Question Mark
    '!',       # Exclamation Mark
    ':',       # Colon
    ';',       # Semicolon
    "'",       # Apostrophe
    '"',       # Double Quotation Mark
    "'",       # Single Quotation Mark (also listed above as Apostrophe)
    '(',       # Left Parenthesis
    ')',       # Right Parenthesis
    '[',       # Left Square Bracket
    ']',       # Right Square Bracket
    '{',       # Left Curly Brace
    '}',       # Right Curly Brace
    '—',       # Em Dash
    '-',       # En Dash and Hyphen
    '...',     # Ellipsis
    '/',       # Slash
    '\\',      # Backslash
    '|',       # Pipe
    '@',       # At Symbol
    '#',       # Hash
    '&',       # Ampersand
    '*',       # Asterisk
    '+',       # Plus Sign
    '=',       # Equals Sign
    '~',       # Tilde
    '`'        # Backtick
]
count = 0
for letter in snowball:
  if letter.isupper() or letter.islower() or letter.isspace() or letter in punctuations:
    count += 1
  else:
    pass
print(count)
```
